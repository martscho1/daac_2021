__Q1:__ Open the 'mdna_2019_aapl.txt' file (which contains the text from the Management Discussion and Analysis (MD&A) section of the 2019 annual report of Apple), read its contents and assign it to a variable ('aapl_19'), close the file, and print the text.

````python
# Open the text file, read its contents, close the file, and print the text

with open('data\mdna_2019_aapl.txt', 'r', encoding="utf8") as fhand:
    aapl_19 = fhand.read()
    fhand.close()

print(aapl_19)
````

__Q2:__ Calculate the frequency of all words in the text, save the words and counts as key-value pairs in a dictionary ('counts'), and print the dictionary. Critically examine the resulting dictionary. What kinds of problems do you observe?

````python
# Calculate the word frequency, save as key-value pairs in a dictionary, and print the dictionary

counts = dict()
for word in aapl_19.split():
    if word not in counts:
        counts[word] = 1
    else:
        counts[word] = counts[word] + 1

print(counts)
````

__Q3:__ In order to solve these problems, you decide to first preprocess the text data, and only then calculate the frequency of all words in the text. As a first preprocessing step, convert the text data to lowercase and assign it to a new variable ('aapl_19_alt'), and print it.

````python
# Standardize case

aapl_19_alt = aapl_19.lower()
print(aapl_19_alt)
````

__Q4:__ As a second preprocessing step, remove the numbers from the text, and print it. Suppose that you want to keep the word '10-k' in the text, how can you recover this word?

````python
# Remove numbers

aapl_19_alt = ''.join([i for i in aapl_19_alt if not i.isdigit()])

# Recover the word '10-k'

aapl_19_alt = aapl_19_alt.replace('-k', '10-k')

print(aapl_19_alt)
````

__Q5:__ As a third preprocessing step, remove the punctuation and special characters from the text, and print it. Critically examine the result of this operation. Are you happy with the result? How could you have improved the result?

````python
# Remove punctuation and special characters

aapl_19_alt = aapl_19_alt.translate(aapl_19_alt.maketrans('', '', string.punctuation))

# Add characters to string.punctuation

string.punctuation = string.punctuation + '“®€”'

print(aapl_19_alt)
````

__Q6:__ As a fourth preprocessing step, remove the stopwords (using the stopwords list from the NLTK library) from the text, and print it.

````python
# Remove stopwords

from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')

aapl_19_alt = ' '.join([word for word in aapl_19_alt.split() if word.lower() not in sw_nltk])
print(aapl_19_alt)
````

__Q7:__ As a fifth preprocessing step, word tokenize the text, and print it.

````python
# Tokenize text

aapl_19_alt = aapl_19_alt.split()
print(aapl_19_alt)
````

__Q8:__ Calculate the frequency of all words in the text, save the words and counts as key-value pairs in a dictionary ('counts'), and print the dictionary.

````python
# Calculate the word frequency, save as key-value pairs in a dictionary, and print the dictionary

counts = dict()
for word in aapl_19_alt:
    if word not in counts:
        counts[word] = 1
    else:
        counts[word] = counts[word] + 1

print(counts)
````

__Q9:__ Provide a bar chart showing the word frequency for the 50 words with the highest frequencies.

````python
# Bar chart showing the word frequencies

lst = list()
for key, value in list(counts.items()):
    lst.append((int(value), key))

lst.sort(reverse = True)
lst = lst[:50]

lst = [(k, v) for (v, k) in lst]

pd.DataFrame(lst, columns=['label','value']).set_index('label').plot(kind='bar');
````

__Q11:__ Count the total number of times that the words from the 'competition_words' wordlist occur in the text ('count'), and print it. Also, think about how you would proceed if you instead want to count the number of words from the 'competition_words' wordlist that occur at least once in the text.

````python
# Calculate the total number of times that the words from the 'competition_words' wordlist occur in the text

count = 0
for word in aapl_19_alt:
    if word in competition_words:
        count += 1

print(count)
````
