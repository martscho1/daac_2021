__Q3:__ Perform the following operations on the DataFrame: (a) select only the following six variables ('id', 'brand', 'reviews.date', 'reviews.doRecommend', 'reviews.rating', 'reviews.text'), (b) drop all rows with missing values, and (c) use the `info()` method to display general information about the variables.

__Q7:__ Define a custom function ('get_year') that uses the `split()` method to split a string on the '-' character and then selects the part of the string in front of the first appearance of this character. Then create a new variable ('reviews.year') by applying the custom function (using the `apply()` method) to the 'reviews.date' variable, and display the first five rows of the DataFrame.

````python
# Define the function and apply it

def get_year(text):
    return text.split('-')[0]

reviews_df['reviews.year'] = reviews_df['reviews.date'].apply(get_year)
reviews_df.head()
````

__Q8:__ Select all instances where the 'reviews.year' variable is equal to either 2015, 2016 or 2017, and then display the first five rows of the DataFrame.

````python
# Select the instances

reviews_df = reviews_df[(reviews_df['reviews.year'] == '2015') |
                        (reviews_df['reviews.year'] == '2016') |
                        (reviews_df['reviews.year'] == '2017')]
reviews_df.head()
````

__Q9:__ Perform the following operations on the 'reviews.text' variable: (a) use the `lower()` method to change the case of all characters to lowercase, and (b) use the `strip()` method to remove any leading and/or trailing spaces. (Hint: don't forget to put `str.` in front of the `lower()` and `strip()` methods.)

````python
# Use the lower() and () strip() methods

reviews_df['reviews.text'] = reviews_df['reviews.text'].str.lower()
reviews_df['reviews.text'] = reviews_df['reviews.text'].str.strip()
````

__Q10:__ Use the following code to import the string module:
```python
import string
```
Then define a custom function ('remove_punctuation') that replaces all characters in a string that are included in `string.punctuation` by an empty space (''), and apply this custom function (using the `apply()` method) to the 'reviews.text' variable, and display the first five rows of the DataFrame.

````python
# Define the function and apply it

import string

def remove_punctuation(text):
    for character in string.punctuation:
        text = text.replace(character, '')
    return text

reviews_df['reviews.text'] = reviews_df['reviews.text'].apply(remove_punctuation)
reviews_df.head()
````

__Q11:__ Remove all digits from the 'reviews.text' column by applying (using the `apply()` method) the following `lambda` function (`lambda` x: " ".join(x `for` x `in` x.split() `if` `not` x.isdigit())) to the 'reviews.text' variable, and display the first five rows of the DataFrame.

````python
# Remove the digits

reviews_df['reviews.text'] = reviews_df['reviews.text'].apply(lambda x: " ".join(x for x in x.split() if not x.isdigit()))
reviews_df.head()
````

__Q12:__ Create a new variable ('reviews.words') by using the `split()` method on the 'reviews.text' column, and then display the variables 'reviews.text' and 'reviews.words'. (Hint: don't forget to put `str.` in front of the `split()` method.)

````python
# Use the split() method

reviews_df['reviews.words'] = reviews_df['reviews.text'].str.split()
reviews_df[['reviews.text', 'reviews.words']]
````

__Q13:__ Use the following code to import the 'stopwords' list from the NLTK library:
```python
from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')
```
Then define a custom function ('remove_stopwords') that removes all words from a list of words that are included in the NLTK stopwords list. Next, apply this custom function (using the `apply()` method) to the 'reviews.words' variable to remove all stopwords, and display the 'reviews.text' and 'reviews.words' variables. __(Please be patient! It may take some time for this code to finish.)__

````python
# Define the function and apply it

from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')

def remove_stopwords(word_list):
    return [w for w in word_list if w not in sw_nltk]

reviews_df['reviews.words'] = reviews_df['reviews.words'].apply(remove_stopwords)
reviews_df[['reviews.text', 'reviews.words']]
````

__Q14:__ Create a new DataFrame ('negative_reviews_df') that contains all instances for which the 'reviews.rating' variable is equal to 1. Next, create another new DataFrame ('positive_reviews_df') that contains all instances for which the 'reviews.rating' variable is equal to 5.

````python
# Create the two DataFrames

negative_reviews_df = reviews_df.loc[reviews_df['reviews.rating'] == 1]
positive_reviews_df = reviews_df.loc[(reviews_df['reviews.rating'] == 5)]
````

__Q15:__ Use the `Series()` method to create a Series that consists of the number of appearances (calculated by using the `sum()` method) of the words in the 'reviews.words' variable in the 'negative_reviews_df' DataFrame.
````python
# Create the Series and display the ten most often appearing words

negative_words = pd.Series(negative_reviews_df['reviews.words'].sum())
negative_words.value_counts().head(10)
````
