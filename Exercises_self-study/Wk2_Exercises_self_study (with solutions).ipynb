__Q1:__ Use the `read_excel()` method to import the first worksheet ('data') of the Excel file 'sommerville.xlsx' as a pandas DataFrame, and display the first five rows of the DataFrame. Next, in a separate cell, display general information about the dataset using the `info()` method.

__NOTE:__ The dataset for this (and the following) question(s), which contains transaction data from the City of Somerville (Massachusetts), is taken from the following website: https://accountingisanalytics.com.

````python
# Import the data

df_data1 = pd.read_excel('data/somerville.xlsx', sheet_name='data')
df_data1.head()
````

````python
# Display general information

df_data1.info()
````

__Q2:__ Use the `columns` attribute to display the column labels of the dataset. If there are any column labels with leading or trailing spaces, use the `str.strip()` method to remove them, and display the column labels once more.

````python
# Display (and remove spaces from) the column names

print(df_data1.columns.tolist())
df_data1.columns = df_data1.columns.str.strip()
print(df_data1.columns.tolist())
````

__Q3:__ As you will have seen when using the `info()` method above, the data type of the 'Amount' variable is 'object'. Given that we want to do calculations with this variable, we need to convert it to integer (or float). However, doing so using the `astype()` method:

```python
df_data1['Amount'] = df_data1['Amount'].astype(int)
```

leads to the following error (Try it!):
```python
ValueError: invalid literal for int() with base 10: '  $-    '
```

So appearently there are rows that start with `'  $-'`. Use the `str.startswith()` method to identify and display those rows.

````python
# Display the rows

df_data1[df_data1['Amount'].astype(str).str.startswith('  $-')]
````

__Q4:__ Use the `~` sign and the `str.startswith()` method to remove the rows that start with `'  $-'`, and display the first five rows of the DataFrame.

````python
# Drop the rows

df_data1 = df_data1[~df_data1['Amount'].astype(str).str.startswith('  $-')]
df_data1.head()
````

__Q5:__ Use the `astype()` method to convert the data type of the 'Amount' variable to integer, and use the `info()` method to display general information about the variables. The data type of the 'Amount' variable should now be 'int64' (or 'int32').

````python
# Convert the data type, and display general information

df_data1['Amount'] = df_data1['Amount'].astype(int)
df_data1.info()
````

__Q6:__ Use the following code to create a new variable ('year'):

```python
df_data1['year'] = df_data1.loc[:,'Check Date'].astype(str).apply([lambda x: x.split('-')[0]])
```

Then use the `groupby()` method to calculate the total amount per year, and display the results.

````python
# Create the new variable ('year'), and perform the calculation

df_data1['year'] = df_data1.loc[:,'Check Date'].astype(str).apply([lambda x: x.split('-')[0]])
df_data1.groupby('year')['Amount'].sum()
````

__Q7:__ Use the `groupby()` method to calculate the total amount per department for the four-year period (2013-2016), and display the results.

````python
# Perform the calculation

df_data1.groupby('Department')['Amount'].sum()
````

__Q8:__ Use the `groupby()` method to calculate the average amount per category ('Category of Gov') for each of the four years, rounded to two digits, and display the results.

````python
# Perform the calculation

df_data1.groupby(['year', 'Category of Gov'])['Amount'].mean().round(2)
````

__Q9:__ Use the following code to create a new DataFrame based on the current DataFrame:

__Q10:__ Use the `read_excel()` method to import the second worksheet ('budget') of the Excel file 'sommerville.xlsx' as a pandas DataFrame, and display the first five rows of the DataFrame. (Hint: You may have to skip rows at the top and/or bottom of the worksheet when importing it!)

````python
# Import the data

df_budget = pd.read_excel('data/somerville.xlsx', sheet_name='budget', skiprows=4, skipfooter=1)
df_budget
````

__Q11:__ Merge the 'df_budget' and 'df_data2' DataFrames, such that the 'Amount' variable becomes the right-most column, assign the merged DataFrame to a variable ('df_comb'), then use the `rename()` method to rename the "Variance \n$" and "Variance \n%" column names (where you remove the '\n'), and display its contents.

````python
# Merge the DataFrames, and rename the columns

df_comb = pd.merge(df_budget, df_data2, how="outer", left_on="Expenditure (Item Class)", right_on="Item Class")
df_comb.rename(columns={"Variance \n$": "Variance $", "Variance \n%": "Variance %"}, inplace=True)
df_comb
````

__Q12:__ Calculate and fill in the values of the columns with missing values ('NaN'), (where applicable) rounded to two digits, and display the DataFrame.

````python
# Fill in the columns with missing values

df_comb['Actual'] = df_comb['Amount']
df_comb['Variance $'] = df_comb['Actual'] - df_comb['Budget']
df_comb['Variance %'] = ((df_comb['Variance $'] / df_comb['Budget']) * 100).round(2)
df_comb
````

__Q13:__ Use the `drop()` method to remove the 'Amount' column from the DataFrame, and display its contents.

````python
# Drop the column

df_comb = df_comb.drop(['Amount'], axis=1)
df_comb
````

__Q14:__ Create a new variable ('Actual_perc') that is equal to the actual amount per item class divided by the sum of all actual amounts, rounded to two digits, and display the DataFrame.

````python
# Create the new variable ('Actual_perc')

df_comb['Actual_perc'] = ((df_comb['Actual'] / df_comb['Actual'].sum()) * 100).round(2)
df_comb
````

__Q15:__ Create a new variable ('Budget_perc') that is equal to the budget amount per item class divided by the sum of all budget amounts, rounded to two digits, and display the DataFrame.

````python
# Create the new variable ('Budget_perc')

df_comb['Budget_perc'] = ((df_comb['Budget'] / df_comb['Budget'].sum()) * 100).round(2)
df_comb
````

__Q16:__ Create a new variable ('Diff_perc') that is equal to the difference between the actual and budget percentages per item class, rounded to two digits, and display the DataFrame.

````python
# Create the new variable ('Diff_perc')

df_comb['Diff_perc'] = df_comb['Actual_perc'] - df_comb['Budget_perc']
df_comb
````
